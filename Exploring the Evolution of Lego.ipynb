```python
# Question 1: What percentage of all licensed sets ever released were Star Wars themed? Save your answer as a variable the_force in the form of an integer (e.g. 25).

#Question 2: In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)? Save your answer as a variable new_era in the form of an integer (e.g. 2012).
```

```python
# Use this cell to begin your analyses, and add as many cells as you would like!

import pandas as pd

#loading in data sets and droping rows with NaN in the set_num column from lego_sets database
df1 = pd.read_csv('datasets/lego_sets.csv').dropna(subset=['set_num'])

df2 = pd.read_csv('datasets/parent_themes.csv')
```

```python
#subsetting all themes that are licesened 
licensed_themes = df2[df2['is_licensed']==True]

#creating a list of licensed themes
themes_list = licensed_themes['name'].tolist()

#obtaining all rows that have licensed themes in df1
licensed_sets = df1[df1["parent_theme"].isin(themes_list)]

#counting number of times 'Star Wars' occurs in the 'parent_theme column
#diving by total number of rows in licensed_sets
#multiplying by 100 to convert to percentage
the_force = (licensed_sets['parent_theme'].value_counts()[0]/licensed_sets.count()[0])*100

the_force
```

```python
#getting the year where Star Wars was not the most pop licensed theme
#grouping the dataset by 'year' and extracting the 'parent_theme' column
#finding the most commonly observed value/theme in each year by finding the mode
licensed_sets = licensed_sets.groupby('year')['parent_theme'].agg(pd.Series.mode)

#creating subset where Star Wars was not the most common and slicing to get the first item
era = licensed_sets[licensed_sets != 'Star Wars'].index[0]

#converting year to a 4 digit integer
new_era = int(era) 

new_era
```
